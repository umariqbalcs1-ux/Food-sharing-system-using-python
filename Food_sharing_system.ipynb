***PYTHON PROJECT (FOOD SHARING SYSTEM)***

***Here is the python code by using all the concepts that are discussed during lectures ***

```
food_inventory=[]
def donate_food():
  print("<<<Donate food>>")
  name=input("Enter the food name : ")
  category=input("Enter the category : ")
  quantity=input("Enter the quantity : ")
  location=input("Enter the loaction : ")
  # Store information in a dictionary
  item = {
        "name": name,
        "quantity": quantity,
        "category": category,
        "location": location
    }
  food_inventory.append(item)
  print("Items added successfully")
def view_food():
  print("<<View food>>")
  if not food_inventory:
    print("No items available")
  else:
    for item in food_inventory:
      print(f"Name : {item['name']}")
      print(f"Quantity : {item['quantity']}")
      print(f"Category : {item['category']}")
      print(f"Location : {item['location']}")
def search_food():
  print("<<Search food>>")
  search_category=input("Enter the category to search : ")
  found=False
  for item in food_inventory:
    if item['category'].lower() == search_category.lower():
      print("<<Found>>")
      found=True
    if not found:
      print("No items found ")
def main():
  while True:
    print("\nFood Sharing System")
    print("Press 1 for . Donate Food")
    print("Press 2 for . View Food")
    print("Press 3 for . Search Food")
    print("Press 4 for . Exit")
    choice = input("Enter your choice : ")
    if choice=='1':
      donate_food()
    elif choice=='2':
      view_food()
    elif choice=='3':
      search_food()
    elif choice=='4':
      print("Thank you for using our system")
      break
    else:
      print("Invalid choice")
if __name__=="__main__":
            main()

```
